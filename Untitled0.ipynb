```python
import shutil
import random
from pathlib import Path
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import zipfile

# ---------------- CONFIG ----------------
DATA_ROOT = Path("archive")      # folder OR zip without extension
ZIP_FILE = Path("archive.zip")   # optional
OUTPUT_DIR = Path("dataset_cls")
TRAIN_RATIO = 0.7
VAL_RATIO = 0.2
SEED = 42
IMG_EXT = ".jpg"                 # change if png/jpeg
# ----------------------------------------

random.seed(SEED)

def extract_if_needed():
    if DATA_ROOT.exists():
        print("Dataset folder found. Skipping extraction.")
        return

    if ZIP_FILE.exists():
        print("Extracting zip file...")
        with zipfile.ZipFile(ZIP_FILE, 'r') as z:
            z.extractall(DATA_ROOT)
    else:
        raise FileNotFoundError("No dataset folder or zip file found.")

def load_samples():
    images_dir = DATA_ROOT / "images"
    labels_dir = DATA_ROOT / "labels"

    if not images_dir.exists() or not labels_dir.exists():
        raise RuntimeError("Expected 'images/' and 'labels/' folders")

    samples = []
    for label_file in labels_dir.glob("*.txt"):
        image_file = images_dir / f"{label_file.stem}{IMG_EXT}"
        if image_file.exists():
            class_id = label_file.read_text().strip()
            samples.append((image_file, class_id))

    if not samples:
        raise RuntimeError("No valid image-label pairs found")

    return samples

def prepare_folders(class_ids):
    for split in ["train", "val", "test"]:
        for cid in class_ids:
            (OUTPUT_DIR / split / cid).mkdir(parents=True, exist_ok=True)

def split_and_copy(samples):
    images, labels = zip(*samples)

    X_train, X_tmp, y_train, y_tmp = train_test_split(
        images, labels,
        train_size=TRAIN_RATIO,
        stratify=labels,
        random_state=SEED
    )

    val_ratio = VAL_RATIO / (1 - TRAIN_RATIO)
    X_val, X_test, y_val, y_test = train_test_split(
        X_tmp, y_tmp,
        train_size=val_ratio,
        stratify=y_tmp,
        random_state=SEED
    )

    splits = {
        "train": (X_train, y_train),
        "val": (X_val, y_val),
        "test": (X_test, y_test),
    }

    for split, (imgs, lbls) in splits.items():
        for img, lbl in tqdm(zip(imgs, lbls), total=len(imgs), desc=split):
            dest = OUTPUT_DIR / split / lbl / img.name
            shutil.copy(img, dest)

def main():
    extract_if_needed()
    samples = load_samples()
    class_ids = sorted(set(lbl for _, lbl in samples))
    prepare_folders(class_ids)
    split_and_copy(samples)

    print("Dataset ready for YOLO Classification")
    print("Classes:", class_ids)
    print("Total samples:", len(samples))

if __name__ == "__main__":
    main()
```

```python
from ultralytics import YOLO

def train():
    model = YOLO("yolov8n-cls.pt")

    model.train(
        data="dataset_cls",
        epochs=50,
        imgsz=224,
        batch=32,
        optimizer="Adam",
        lr0=1e-3,
        device=0,              # use 'cpu' if no GPU
        project="runs_cls",
        name="custom_cls",
        pretrained=True
    )

if __name__ == "__main__":
    train()
```
